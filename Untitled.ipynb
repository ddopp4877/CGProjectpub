```python
def findInListIdx(myList,key):
    for i in range(0,len(myList)):
        if str(key) == str(myList[i]):
            return i
    
```

```python
allCount = []
majorList = list(set(Roster['Acad Plan']))

for major in majorList:
    allCount.append(len(np.where(Roster['Acad Plan'] == str(major))[0]))
  
```

```python
diffCount = []
for major in majorListDrop:
    if allCount[findInListIdx(majorList,major)] == 0:
        diffCount.append(0)
    
    else:
        diffCount.append(majorCountDrop[findInListIdx(majorListDrop,major)]/(allCount[findInListIdx(majorList,major)]))
```

```python
plt.figure(figsize=(10,10))
plt.scatter(diffCount,np.arange(0,len(majorListDrop)))
plt.yticks(np.arange(0,len(majorListDrop)),majorListDrop)
plt.xticks(np.round(np.arange(0,1,1/len(majorCountDrop)),2),np.round(np.arange(0,1,1/len(majorCountDrop)),2),rotation=45)
plt.hlines(np.arange(0,len(majorListDrop)),xmin = 0,xmax = 1)
plt.xlabel('dropout ratio')
plt.show()
```
